# 지옥이 필요한 세 가지 이유
----------------------------
## 세가지 관점
>- __정의__의 관점
>- __하나님의 성품__의 관점
>- __인간의지 행위__에 대한 관점
----------------------------
### 1. 정의의 관점

>- 인간: 쾌락을 선택하고 고통을 피하는 합리적인 동물
>- 형벌은 각 범죄에 부과 -->  __범죄의 쾌락 < 범죄의 고통__
>- 국가: 국민의 생명과 재산을 보호하기 위해 범죄자들에 대한 처벌을 형법에 규정한다.
>- 범죄에 비례하여 형벌은 결정된다.
>- 공리주의자 벤담: 형벌의 가장 큰 목적 -->  범죄 예방
>- 현실적으로는 범죄가 완전히 예방되기는 불가능하며 범죄자에 대한 처벌도 완전하지 않다.
>- 죄에 대한 하나님의 완전한 사후의 처벌이 없다면 __완전한 정의__ 는 없는 것
------------------------------
### 2. 하나님의 성품

>- __거룩한 하나님__ 은 어떤 죄도 묵과하지 않는다.
>- 죄를 지으면서도 하나님의 자비와 용서를 비방하는 인간들의 __부패한 본성__ 때문에 지옥이 필요하다
>- 지옥은 사랑과 공의로 나타나는 __하나님의 거룩함의 속성__ 에서도 반드시 필요하다.
>- 범죄자를 처벌하지 않는 국가가 __정의로운__ 나라가 될 수 없음 = 죄를 방관하고 죄인의 악행을 묵인하는 신은 __정의의__ 하나님이 될 수 없음
--------------------------------
### 3. 인간의 의지적 선택의 관점

>- 인간은 도구적 가치가 아니라 __본래적 가치__ 를 가지고 있다. 
>- 인간이 도구나 수단이 아니라 그 자체로 __목적적인 존엄성__ 을 가진 존재라는 의미이다.
>- 하나님은 인간을 __자유의지__ 를 가지고 행동하고 __선택__ 을 할 수 있는 존재로 창조하셨다.
>- 하나님은 인간을 강압하지 않고 __의지적 선택__ 을 존중하신다.
>- 천국행과 지옥행도 인간의 __선택__ 이 변수로 작용한다.
>- "누구든지 주의 이름을 부르는 자"는 천국에 가지만, "하나님의 아들을 믿지 않는 자"는 지옥에 간다
>- 지옥의 본질은 __'하나님과의 파괴된 관계'__ 로 설명할 수 있다.
>- 예수 그리스도를 영접하는 사람은 하나님의 자녀과 되며 구원을 받는다.
>- 지옥이란 사랑의 하나님으로부터 완전히 분리되는 __영원한 추방__ 이다
>- 지옥은 다시는 번복되거나 변경할 수 없는 __영원한 최종선고__ 이다. 하나님을 무시하고 조롱하는 삶을 산 이는 다시는 하나님의 형상을 회복할 수 없는 __영원한 형벌__ 을 받기 때문이다.
>- 지옥은 인간의 __본래적 가치__ 의 관점에서 설명이 가능하다.
>- 영원한 시간을 '하나님의 나라'와 '지옥'중에 어느 곳에서 보낼지는 우리의 __의지__ 이며 인격적인 __선택__ 여하에 따라 달라지게 된다.
>- 독생자를 보내시면서까지 모든 인류를 구원하고자 하셨던 하나님의 사랑이 모든 사람들에게 동일한 결과를 가져오지 못하는 것도 __'자유의지의 남용'__ 때문이다. 
>- 인류를 향한 하나님의 목표는 지옥이 아니라 __천국__ 이다.

# 지옥의 영원한 형벌은 지나친 가중처벌이 아닌가요?
---------------------------------
## 1. 범죄에 대한 처벌로 교도소를 다녀온다고 해서 지은 죄가 사라지는가?

>- 지옥에서 고통을 당한다고 해서 죄 자체가 사라졌으니 지옥형벌에서 벗어날 __권리__ 가 있다고 말할 수 있을까? 
>- 더글라스: 형량이 __죄__ 자체를 소멸하게 하지 않는다
>- 모든 사람은 __죄인__ 이다.
>- 죄를 용서받는 궁극적인 길은 예수 그리스도의 __십자가의 희생__ 을 통하는 길 밖에 없다.
>- 예수님은 마땅히 처벌받아야 하는 죄인들의 죄를 용서하기 위해서 자신을 __대속물__ 로 주셔서 __하나님의 공의__ 를 충족시켰다.
----------------------------------
## 2. 지옥의 형벌의 *영원성* 에 대해 생각하자

>- 지옥은 하나님과의 __영원한 격리__ 를 선고받은 죄인들이 가는 곳이다.  
>- C.S루이스: 지옥이 과잉처벌은 영원을 __3차원적 시간의 연장__ 으로 보기 때문에 생기는 오해이다. 지옥의 __형벌 영원성__ 은 우리가 알고 있는 3차원적 시간이 아니라 __완전해진__ 시간개념 속에서의 형벌을 의미한다.

----------------------------
## 3. 형사처벌로 인한 형량은 범죄행위에 소요된 *시간* 이 아니라, 범죄행위 자체의 *잔인성* 의 정도와 *범죄대상* 에 따라 달라진다. 

>- 단순히 범죄행위에 소요된 __시간__ 에 비례해서 형벌을 정하지는 않는다.
>- 흉악한 살인 사건도 범죄에 소요된 시간은 실제로 얼마 안 될 수 있다.
>- 인간이 지을 수 있는 가장 나쁜 범죄는 창조주 __하나님__ 을 우롱하고, __예수님__ 을 모욕하며, __성령 하나님__ 을 비방하는 것이다.
>- 영원한 하나님을 거역하였기 때문에 __영원한 형벌__ 을 받는 것이 마땅하다.

# 지옥심판보다는 윤회가 더 나은 생각 아닌가요?

## 윤회사상의 오류
--------------------------------------
### 1. 모든 사람이 지옥에서 벌을 받으면 다 *회개할 수 있다는 전제* 를 가지고 있다.

>- 한번 죽으면 __윤회__ 나 __환생__ 을 할 수 없다. 
>- 윤회의 과정을 통해 개선될 수 있었다면 지금 극히 발달했어야 할 텐데 현실은 그렇지 않다.
>- __시간__의 역사가 흘러갈수록 도덕이 발달한다는 결론은 얻을 수 없다.
-------------------------------------
### 2. 사람들과 짐승들은 기본적으로 모두 자기를 *보호* 하고 *유지* 하고자 하는 본성을 가지고 있다.

>- 불교와 힌두교는 __존재의 소멸__ 로 윤회를 벗어날 수 있다고 말한다.
>- 인간은 영혼과 육체를 가진 __단일한__ 개체로서 영원히 사는 __인격적인__ 존재이다. 
>- __무아__ 가 된다는 것은 자기를 보존하는 __내재적 본성__ 과도 배치가 된다.
>- 무아는 __죽음__ 의 다른 표현에 지나지 않으며 고통의 문제를 이런 식으로 해결하는 것은 올바르지 않아 보인다. 
--------------------------
### 3. 인간의 몸이 다른 *비인간적 존재* 로 환생한다는 것은 자기모순이다

>- __유전적 구조__ 가 달라서 결코 동일인이 될 수 없다.

### 4. 윤회에 대한 증거로 전생에 대한 기억과 정신적 기질 패턴의 *심리학적 연속성* 을 고려해보자

>- 전생에 대한 어떤 __기억__ 도 갖고 있지 않다는 점을 고려하면 전생에 대한 __기억__ 은 윤회의 증거로 사용될 수 없다.
>- 전생에 대한 기억은 인류의 __보편적인 경험__ 이 아니다. 
--------------------------------
### 5. 윤회가설은 '전생..그 전생'으로 *무한 소급* 을 해야 하는데, *최초의 생애* 를 상정하지 못한다.

>- 윤회가설이 주장하는 전생의 이론은 개인 혹은 인류의 __기원__ 을 전혀 설명할 수 없다.
