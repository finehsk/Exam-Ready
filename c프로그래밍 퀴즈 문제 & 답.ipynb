```
1. 해석
1.1 C언어에서는 데이터를 _stream_를 통해서 읽고 쓴다. *헷깔림
1.2 파일에 접근하기 위해서, 우리는 반드시 파일을 스트림과  _associate_ 해야한다. *헷깔림
1.3 _text_스트림: 문자의 연속으로 이루어져있다. 
1.4 _binary_스트림: 데이터 값의 연속으로 이루어져있다. 

2. 이유
2.1 스트림이란 무엇이기에 파일을 읽고 쓸 수 있지? 스트림은 파일 처리 할 때의 데이터 경로이다. 터미널 또는 cmd와 파일 사이를 연결하는  빨대같은 것이다. 프로그램에 의해서 만들어지는 독립체이다(?)
2.2 associate: 접근한다는 의미. access와 비슷한 느낌이다. 스트림으로 파일을 읽어오는 과정을 'associate'한다고 표현한다. 
2.4 데이터 값은 0과 1의 숫자고 이해할 것.
```
